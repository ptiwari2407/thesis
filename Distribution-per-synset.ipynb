```python
import numpy as np
import pandas as pd
```

```python
file = 'data/syn_dict.json'
import json
data = json.load(open(file))
```

```python
syn_length = []
for elem in data:
    syn_length.append(len(data[elem]))
```

```python
arr = np.array(syn_length)
(unique, counts) = np.unique(arr, return_counts=True)
frequencies = np.asarray((unique, counts)).T
print(frequencies)
```
